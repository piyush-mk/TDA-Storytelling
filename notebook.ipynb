```python
# frequency distribution of food categories
food_columns = ["Epic-1", "Satirical-1", "Lyric-Rom-1", "Epic-2", "Didactic-1", 
                "Idyllic-1", "Satirical-2", "Lyric-Sad-1", "Didactic-2", 
                "Idyllic-2", "Lyric-Sad-2", "Lyric-Rom-2"]
food_counts = df[food_columns].apply(pd.Series.value_counts)
print(food_counts)
```

## Proportion of food categories

```python
food_counts = df[food_columns].apply(pd.Series.value_counts, normalize=True)
food_counts = food_counts.loc[1, :]


plt.figure(figsize=(8, 8))
plt.pie(food_counts, labels=food_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Proportion of food categories overall")
plt.show()
```

## Frequency distribution of food categories

```python
food_columns = df.columns[:-2]
food_counts = df[food_columns].stack().value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=food_counts.index, y=food_counts.values)
plt.title("Frequency distribution of food categories")
plt.xlabel("Food categories")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, cmap="coolwarm", annot=True)
plt.title("Correlation matrix")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["Age"])
plt.title("Distribution of age")
plt.xlabel("Age")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="Age", y="Epic-1", data=df)
plt.title("Age vs. food category - Epic-1")
plt.xlabel("Age")
plt.ylabel("Food category")
plt.show()
```

```python
food_columns = ["Epic-1", "Satirical-1", "Lyric-Rom-1", "Epic-2", "Didactic-1", 
                "Idyllic-1", "Satirical-2", "Lyric-Sad-1", "Didactic-2", 
                "Idyllic-2", "Lyric-Sad-2", "Lyric-Rom-2"]
food_counts = df[food_columns].apply(pd.Series.value_counts, normalize=True)
food_counts = food_counts.loc[1, :]
food_counts_by_gender = df.groupby("Gender")[food_columns].apply(lambda x: x.eq(1).sum() / len(x))
food_counts_by_gender = food_counts_by_gender.reset_index().melt(id_vars="Gender", var_name="Food Category", value_name="Proportion")
plt.figure(figsize=(10, 6))
sns.barplot(x="Food Category", y="Proportion", hue="Gender", data=food_counts_by_gender)
plt.title("Proportion of food categories by gender")
plt.xlabel("Food categories")
plt.ylabel("Proportion")
plt.show()
```

## proportion of people who selected each food category by gender

```python
food_counts_by_gender = df.groupby(["Gender"])[food_columns].sum() / df.groupby(["Gender"])[food_columns].count()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
fig.suptitle("Proportion of food categories by gender")

ax1.pie(food_counts_by_gender.loc[1, :], labels=food_counts.index, autopct="%1.1f%%", startangle=90)
ax1.set_title("Females")

ax2.pie(food_counts_by_gender.loc[2, :], labels=food_counts.index, autopct="%1.1f%%", startangle=90)
ax2.set_title("Males")

plt.show()
```

## bar chart of the food preferences for the first music genre played

```python
food_genre1 = ['Epic-1', 'Satirical-1', 'Lyric-Rom-1', 'Didactic-1', 'Idyllic-1', 'Lyric-Sad-1']
food_counts_genre1 = df[food_genre1].stack().value_counts().sort_index()
sns.barplot(x=food_counts_genre1.index, y=food_counts_genre1.values)
plt.xticks(rotation=45)
plt.xlabel('Food Type')
plt.ylabel('Count')
plt.title('Food Preferences for First Music Genre Played')
plt.show()
```

## scatter plot of age vs. the number of fast food preferences for the second music genre played


```python
sns.scatterplot(data=df, x='Age', y='Satirical-2', hue='Gender')
plt.xlabel('Age')
plt.ylabel('Count of Fast Food Preferences for Second Music Genre Played')
plt.title('Age vs. Fast Food Preferences for Second Music Genre Played')
plt.show()
```

## food preference counts for each music genre

```python
genre_food_counts = pd.DataFrame({
    'Epic': df[['Epic-1', 'Epic-2']].stack().value_counts(),
    'Satirical': df[['Satirical-1', 'Satirical-2']].stack().value_counts(),
    'Lyric-Rom': df[['Lyric-Rom-1', 'Lyric-Rom-2']].stack().value_counts(),
    'Didactic': df[['Didactic-1', 'Didactic-2']].stack().value_counts(),
    'Idyllic': df[['Idyllic-1', 'Idyllic-2']].stack().value_counts(),
    'Lyric-Sad': df[['Lyric-Sad-1', 'Lyric-Sad-2']].stack().value_counts()
})

genre_food_props = genre_food_counts.div(genre_food_counts.sum(axis=0), axis=1)

genre_food_props.plot(kind='bar', stacked=True)
plt.xticks(rotation=0)
plt.xlabel('Food Type')
plt.ylabel('Proportion')
plt.title('Food Preferences for Each Music Genre')
plt.show()
```

## average food preference for each age group and gender
#### (darker colors indicate stronger preferences)

```python
age_gender_food = df.groupby(['Age', 'Gender']).mean().iloc[:, :-2]

sns.heatmap(age_gender_food, cmap='coolwarm', annot=True)
plt.title('Average Food Preferences by Age and Gender')
plt.show()
```

## scatter plot matrix of the food preferences for all six music genres played

```python
food_genres = ['Epic-1', 'Satirical-1', 'Lyric-Rom-1', 'Didactic-1', 'Idyllic-1', 'Lyric-Sad-1', 
               'Epic-2', 'Satirical-2', 'Lyric-Rom-2', 'Didactic-2', 'Idyllic-2', 'Lyric-Sad-2']
sns.pairplot(data=df, vars=food_genres)
plt.show()
```

## bar chart of the most common food preferences

```python
food_counts = df[food_genres].stack().value_counts().sort_index()
sns.barplot(x=food_counts.index, y=food_counts.values)
plt.xticks(rotation=45)
plt.xlabel('Food Type')
plt.ylabel('Count')
plt.title('Most Common Food Preferences Across All Music Genres Played')
plt.show()
```
